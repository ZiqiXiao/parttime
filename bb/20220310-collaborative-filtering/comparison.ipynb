```python
from CF import UserCFRecTrad, UserCFRecKmeans
import matplotlib.pyplot as plt
```

```python
k_list = [5, 10, 15, 20, 30, 40, 50]
trad_precision_list = []
trad_recal_list = []

kmeans_precision_list = []
kmeans_recal_list = []

UserCFRec_kmeans = UserCFRecKmeans('ml-latest-small/ratings.csv', 'kmeans_sim.joblib', 32)
UserCFRec_trad = UserCFRecTrad('ml-latest-small/ratings.csv', 'trad_sim.joblib')

for k in k_list:
    precision, recall = UserCFRec_trad.metrics(k=k, nitems=25)
    trad_precision_list.append(precision)
    trad_recal_list.append(recall)

    precision, recall = UserCFRec_kmeans.metrics(k=k, nitems=25)
    kmeans_precision_list.append(precision)
    kmeans_recal_list.append(recall)
```

```python
fig, ax1 = plt.subplots()
plt.grid()
plt.plot(k_list, trad_precision_list, 'r', label = 'trad_precision', marker="o", linestyle="--")
plt.plot(k_list, kmeans_precision_list, 'b', label = 'kmeans_precision', marker="o", linestyle="-")
plt.legend(loc='upper left')
plt.xlabel('k')
plt.ylabel('precision')
plt.title('precision')
plt.show()
```

```python
fig, ax1 = plt.subplots()
plt.grid()
plt.plot(k_list, trad_recal_list, 'r', label = 'trad_recal', marker="o", linestyle="--")
plt.plot(k_list, kmeans_recal_list, 'b', label = 'kmeans_recal', marker="o", linestyle="-")
plt.legend(loc='upper left')
plt.xlabel('k')
plt.ylabel('recall')
plt.title('recall')
plt.show()
```
